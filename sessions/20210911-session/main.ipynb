```python
import pandas as pd
import matplotlib.pyplot as plt
dbPath = "../../data/prima-indians-diabetes.csv"

names = ['Preg', 'Plas', 'Pres', 'Skin',
         'Test', 'Mass', 'Pedi', 'Age', 'Class']

data = pd.read_csv(dbPath, names=names)
pd.set_option('precision', 2)
```

```python
# Funcion head(int amount)
data.head(30)
```

```python
# Funcion shape()
data.shape
```

```python
# Funcion dtypes()
data.dtypes
```

```python
# Funcion describe()
pd.set_option('precision', 1)
data.describe()
```

```python
# Funcion groupby()
data.groupby('Class').size()
```

```python
# Funcion corr(str method)
data.corr(method='pearson')
```

```python
# Funcion skew()
data.skew()
```

```python
# Histogram
fig = plt.figure(figsize=(10,10))

ax = fig.gca()
data.hist(ax = ax, color = 'cyan')
plt.show()
```

```python
# Density
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
data.plot(ax=ax, kind='density', subplots=True, layout=(3, 3))
plt.show()
```

```python
# Boxplot
fig = plt.figure(figsize=(10,10))
ax = fig.gca()
data.plot(ax = ax, kind='box', subplots = True , layout = (3,3))
plt.show()
```
