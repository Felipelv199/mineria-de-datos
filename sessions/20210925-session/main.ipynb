```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
filename = "../../data/iris.data.csv"
names = ["Sepal.length", "Sepal.width", "Petal.length", "Petal.width", "class"]
data = pd.read_csv(filename, names=names)

plt.figure(1)
plt.subplots(figsize=(20,20))

plt.subplot(4,2,1)
sea.boxplot(x="class", y="Sepal.length", data=data)
plt.title("Sepal.length")
plt.grid(True)

plt.subplot(4,2,2)
sea.boxplot(x="class", y="Sepal.width", data=data)
plt.title("Sepal.width")
plt.grid(True)

plt.subplot(4,2,3)
sea.boxplot(x="class", y="Petal.length", data=data)
plt.title("Petal.length")
plt.grid(True)

plt.subplot(4,2,4)
sea.boxplot(x="class", y="Petal.width", data=data)
plt.title("Petal.width")
plt.grid(True)
```
